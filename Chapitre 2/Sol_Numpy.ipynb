1. Calculer le sinus, cosinus et la tangente de 45° arrondi à 2 décimales.<br>
2. Afficher la valeur de pi en radian <br>
3. Résultat attendu<br>



```python
import numpy as np

# Convertis 45 degrés en radians
angle_rad = np.radians(45)

# Calcule sin, cos et tan de cet angle
print("sin(45°) =", round(np.sin(angle_rad),2))
print("cos(45°) =", round(np.cos(angle_rad),2))
print("tan(45°) =", round(np.tan(angle_rad),2))

# Convertis pi radians en degrés
print("π radians en degrés =", np.degrees(np.pi))
```

Calculer l'arcsin et l'arccosinus de 0,5. <br>
Calculer l'arctangente de 1. <br>
Afficher les résultats <br>




```python
import numpy as np

# Calcule l'angle dont le sinus est 0.5
angle_sin = np.arcsin(0.5)

# Calcule l'angle dont le cosinus est 0.5
angle_cos = np.arccos(0.5)

# Calcule l'angle dont la tangente est 1
angle_tan = np.arctan(1)

# Affiche les angles en degrés
print("arcsin(0.5) =", np.degrees(angle_sin), "°")
print("arccos(0.5) =", np.degrees(angle_cos), "°")
print("arctan(1) =", np.degrees(angle_tan), "°")
```

Un étudiant observe une échelle appuyée contre un mur. L’échelle mesure 5 mètres de long et forme un angle de 60° avec le sol<br> Il veut connaître :
- La hauteur à laquelle l’échelle touche le mur (opposé).<br>
- La distance entre le pied de l’échelle et le mur (adjacent). <br>

Résultat <br>


```python
import numpy as np

# Données
longueur_echelle = 5  # en mètres
angle_deg = 60        # en degrés

# Conversion en radians
angle_rad = np.radians(angle_deg)

# Calculs trigonométriques
hauteur_mur = longueur_echelle * np.sin(angle_rad)      # côté opposé
distance_mur = longueur_echelle * np.cos(angle_rad)     # côté adjacent

# Affichage
print("L'échelle mesure", longueur_echelle, "mètres et forme un angle de", angle_deg, "° avec le sol.")
print("Hauteur atteinte sur le mur :", round(hauteur_mur, 2), "m")
print("Distance entre le mur et le pied de l'échelle :", round(distance_mur, 2), "m")
```
